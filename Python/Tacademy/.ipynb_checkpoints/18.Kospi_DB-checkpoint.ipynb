```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
# import seaborn as sns
```

```python
kospi_df = pd.read_csv(
    "./data/kospi.csv",
    parse_dates=[0],
    index_col=0,
)
kospi_df.columns
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(
    kospi_df.index,
    kospi_df.Close,     # TOHLC TOCHL ...
)

ax.set_title("KOSPI 2016")
ax.axhline(y=2050, color="red", linestyle="--")

# DataFrame 은 그대로 유지하지만, 그래프의 x 축 범위 
ax.set_xlim([
    "2016-01-01", "2016-11-08",
])

fig.savefig("kospi-2016.png")
# 삼성전자 의 주가 ( 2016년도 ) :: KRX:005930
```

```python
cosmax_df = pd.read_csv(
    "./data/cosmax.csv",
    parse_dates=[0],
    index_col=0,
)
```

```python
cosmax_df.Close.plot(
    title="COSMAX"
)
```

```python
cosmax_df.Close.plot(
    title="COSMAX"
)

plt.annotate(
    "Thaad",
    xy=("2016-02-24", 130000),
    xytext=("2016-02-24", 170000),
    arrowprops=dict(facecolor="black")
)
```

```python
ax = cosmax_df.Close.plot(
    title="COSMAX"
)

ax.axhline(y=115000, color="red", linestyle="--")

plt.annotate(                    # 6월 24일 브렉시트 
    "Thaad",
    xy=("2016-02-24", 130000),
    xytext=("2016-02-24", 170000),
    arrowprops=dict(facecolor="black")
)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(
    cosmax_df.index,
    cosmax_df.Close,
)
ax1.set_title("COSMAX 2016")
ax1.set_ylabel("Close")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(
    cosmax_df.index,
    cosmax_df.Close,
)
ax1.set_title("COSMAX 2016")
ax1.set_ylabel("Close")

ax2.bar(
    cosmax_df.index,
    cosmax_df.Volume,
)

ax2.set_ylabel("Volume")
ax2.set_xlabel([])
```
