````python
import requests
from bs4 import BeautifulSoup

response = requests.get("http://www.naver.com")
dom = BeautifulSoup(response.text, "html.parser")
````

````python
# select/select_one
# element.attrs ( DICT ), element.attrs.get("title"/"href")
# element.text

[
    rank_element.select_one("a").attrs.get("title")
    for rank_element
    in dom.select("#realrank li")
][:-1]
````

````python
from selenium import webdriver
````

````python
driver = webdriver.PhantomJS()
````

````python
driver.get("http://www.naver.com")
````

````python
# find_element[s]_by_css_selector
````

````python
rank_elements = driver.find_elements_by_css_selector("#realrank li")
````

````python
[
    rank_element.find_element_by_css_selector("a").get_attribute("title")
    for rank_element
    in rank_elements
]

# Naver title=""
````

````python
# driver.find_element[s]_by_css_selector
# element.attrs / element.get_attribute("title")
# element.text / element.text
[
    rank_element.find_element_by_css_selector("a").text.split("\n")[0]
    for rank_element
    in rank_elements
]
# Naver title=""
````
