### <strong>02 박스모델의 속성 2
1. <strong>padding</strong> : 콘텐츠 영역과 테두리의 여백설정
2. <strong>margin</strong> : 요소의 주변여백 설정

```python
%%html
<head>
<link href="style.css" rel="stylesheet" type="text/css">
    <style>                                       
    .content1{border:1px solid black; background:lightyellow; 
              padding:10px;}          /* 네 방향의 패딩 모두 10px */

    .content2{border:1px solid black; background:lightyellow;	
              margin-bottom:10px; margin-left:50px;  /* 마진 : 아래 10px 왼쪽 50px*/
              padding:10px 30px;}                    /* 패딩 : 위아래 10px 좌우 30px */ 

    .content3{border:1px solid black; background:lightyellow;          /* 위에서 시계방향 */
          margin:20px 10px 0 100px;     /* 마진: 위쪽 20px, 오른 10px, 아래 없음, 왼쪽 100px */
          padding:10px 30px 10px 50px;} /* 패딩: 위쪽 10px, 오른 30px, 아래 10px, 왼쪽 50px */
    </style>
</head>

<body>
<div class="container">
    <section>
    <h2>예약 방법 및 요금</h2>    
        <div class="content1">
            <h3>예약방법</h3>
            <p> 070.4529.8743 : 인터넷 전화로 시외전화 요금이 부과되지 않습니다.</p>
        </div>
        <div class="content2">
            <h3>예약방법</h3>
            <p> 070.4529.8743 : 인터넷 전화로 시외전화 요금이 부과되지 않습니다.</p>
        </div>
        <div class="content3">
            <h3> 예약 방법</h3>
            <p> 070.4529.8743 : 인터넷 전화로 시외전화 요금이 부과되지 않습니다.</p>
        </div>	  
    </section> 
</div>
</body></html>
```

## <strong>3 배치와 관련된 블록레벨, 인라인레벨

## <strong>4 여러요소를 원하는 형태로 배치하기
<strong>position</strong> : 원하는 형태로 요소 배치하기

```python
%%html
<style>
#box1{
    position:absolute;    /* absolute(절대좌표) : 겹처도 무관 (left/top 기준) */
    left:30px; top:10px;} /* 수평좌표, 수직좌표 */
#box2{
    position:absolute;
    left:80px; top:40px;} /* 수평좌표, 수직좌표 */
</style>

<div id='box1'> <p>여기가 거기더냐?</p> </div>
<div id='box2'> <p>여기가 거기더냐?</p> </div>
```

```python
%%html
<style>
#box1{
    position:static;        /*  static(정적좌표) : 자식의 크기에 따라 부모 크기가 변경 (left/ top 기준) */
    left:130px; top:10px;}  /* 수평좌표, 수직좌표 */
#box2{
    position:static;
    left:170px; top:40px;}  /* 수평좌표, 수직좌표 */
</style>

<div id='box1'>
    <p>여기가 거기더냐?</p>
</div>
<div id='box2'>
    <p>여기가 거기더냐?</p>
</div>
```

```python
%%html
<style>
#box1{ 
    position:relative;      /*  relative(상대좌표) : 겹치지 않도록 연결배치 (left/top기준) */
    left:130px; top:10px;}  /* 수평좌표, 수직좌표 */
#box2{
    position:relative;
    left:170px; top:40px;}  /* 수평좌표, 수직좌표 */
</style>
<div id='box1'>
    <p>여기가 거기더냐?</p>
</div>
<div id='box2'>
    <p>여기가 거기더냐?</p>
</div>
```
