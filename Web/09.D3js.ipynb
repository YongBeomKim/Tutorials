## <strong>4. Event와 데이터를 연계한 D3 시각화

```python
%%html
<head>
    <script src="http://d3js.org/d3.v3.min.js" charset="utf-8"></script>
    <style>
        svg#myGraph_bt { width: 320px; height: 130px; border: 1px solid black; }
        #myGraph_bt .bar { fill : orange; }
    </style>
</head>
<body>
    <h1> 1. CSV 파일에 따른 그래프 바꾸기 </h1>
    <div>
        <button csv_src1="https://raw.githubusercontent.com/YongBeomKim/Tutorials/master/Web/data/mydata1.csv">mydata1.csv 불러오기</button>
        <button csv_src1="https://raw.githubusercontent.com/YongBeomKim/Tutorials/master/Web/data/mydata2.csv">mydata2.csv 불러오기</button>
        <button csv_src1="https://raw.githubusercontent.com/YongBeomKim/Tutorials/master/Web/data/mydata3.csv">mydata3.csv 불러오기</button>
    </div>
    <svg id="myGraph_bt1"></svg>
    <script>
    d3.selectAll("button")
      .on("click", function(){
            var csvFile = this.getAttribute("csv_src1");    // data-src 속성을 읽어옴(즉, CSV 파일 이름)
            var barElements;                               // 막대그래프의 막대 요소를 저장할 변수
            d3.csv(csvFile, 
                function(error, data){
                    var dataSet = [ ]; 
                    for(var i=0; i<data.length; i++){
                           dataSet.push(data[i]["상품A"]); }  
                        barElements = d3.select("#myGraph_bt1").selectAll("rect").data(dataSet)
                        barElements.enter().append("rect")                  // 객체 그리기 1 : 전체
                                   .attr({"class":"bar", "height":20, "x":0, 
                                          "width": function(d, i){return d;},
                                          "y"    : function(d, i){return i*25},})
                        barElements.attr("width", function(d,i){return d;}) // 객체 그리기 2 : 갱신
            });
        });
    </script>
</body>
```

## <strong>5. 세로형 막대그래프 D3 시각화
http://webinformation.tistory.com/36 : JavaScript 함수

```python
%%html
<head>
<script src="http://d3js.org/d3.v3.min.js" charset="utf-8"></script>
<style>
    svg#myGraph_bar { width: 320px; height: 240px; border: 1px solid black; }
    .bar { fill : orange; }
    .barNum {font-size: 9pt;  text-anchor : middle;}
    .axis text {font-family: sans-serif; font-size: 11px;}
    .axis path,
    .axis line {fill: none; stroke: black;}
    .axis_x line {fill: none; stroke: black;}
    .barName {font-size: 9pt; text-anchor : middle;}
</style>
</head>
<body>
    <h1> 1. 세로형 막대그래프에 눈금 표시 </h1>
    <svg id="myGraph_bar"></svg>
    <script>
    var svgHeight = 240;        // SVG 요소의 높이
    var offsetX = 30;           // X 좌표의 오프셋(어긋남의 정도)
    var offsetY = 20;           // Y 좌표의 오프셋(어긋남의 정도)
    var barElements;
    var dataSet = [120, 70, 175, 80, 220];
    
    barElements = d3.select("#myGraph_bar").selectAll("rect").data(dataSet)
    barElements.enter().append("rect")                              // 데이터 수만큼 rect 요소가 추가됨
        .attr({"class":"bar", "width":20, 
               "height":function(d,i){return d;},
               "x":function(d,i){return i * 25 + offsetX;},
               "y":function(d,i){return svgHeight-d-offsetY;}, })
    barElements.enter().append("text") 
        .attr({"class":"barNum",
               "x" : function(d,i){return i * 25 + 10 + offsetX;},
               "y" : svgHeight-5-offsetY })
        .text(function(d,i){return d;})
    barElements.enter().append("text")                               // 막대의 레이블을 표시
        .attr({"class":"barName",                                    
               "x" : function(d,i){return i * 25 + 10 + offsetX;},   // 막대그래프의 표시 간격을 맞춤
               "y" : svgHeight - offsetY + 15, })
        .text(function(d, i) {return ["A","B","C","D","E"][i];})     // 레이블 이름 반환    

    var yScale = d3.scale.linear().domain([0, 300]).range([300, 0])  // 원래 크기, 실체 출력 크기
    d3.select("#myGraph_bar").append("g")                            // 세로 축 눈금표시
      .attr({"class":"axis", "transform":"translate(30,-80)"})
      .call(d3.svg.axis().scale(yScale).orient("left"),)
    d3.select("#myGraph_bar").append("rect")                         // 가로 축 눈금표시
      .attr({"class":"axis_x", "width":320, "height":1, 
               "transform":"translate("+offsetX+", "+(svgHeight-offsetY)+")"})
    </script>
</body>
```

```python
%%html
<head>
<script src="http://d3js.org/d3.v3.min.js" charset="utf-8"></script>
<style>
    svg#myGraph_bar2 { width: 300px; height: 240px; border: 1px solid black; }
    .bar { fill : orange; }
    .barNum {font-size: 9pt;               text-anchor : middle;}
    .axis text {font-family: sans-serif;   font-size: 11px;}
    .axis path,
    .axis line {fill: none;  stroke: black;}
    .axis_x line {fill: none;  stroke: black;}
    .barName {font-size: 9pt;  text-anchor : middle;}
</style>
</head>
<body>
    <h1> 2. 세로형 막대그래프 순서대로 애니메이션 효과 </h1>
    <svg id="myGraph_bar2"></svg>
    <script>
    var svgWidth = 320;
    var svgHeight = 240;
    var offsetX = 30;
    var offsetY = 20;
    var barElements;
    var dataSet = [120, 70, 175, 80, 220, 40, 180, 70, 90];
    var dataMax = 300;
    var barWidth = 20;
    var barMargin = 5;
    
    barElements = d3.select("#myGraph_bar2").selectAll("rect").data(dataSet)
    barElements.enter().append("rect")
        .attr({"class":"bar", "height":0, "width":barWidth,
               "x":function(data,index){return index*(barWidth+barMargin) + offsetX;},
               "y":svgHeight-offsetY,})
        .transition().duration(1000)                // 1초동안 애니메이션 처리
        .delay(function(d,i){return i * 100;})      // 0.1초 대기
        .attr({"y":function(d,i){return svgHeight-d-offsetY;},
               "height":function(d,i){return d;},}) // 애니메이션 대상 속성들
    
    barElements.enter().append("text")              // text 요소 지정
        .text(function(d, i){return d;})
        .attr({"class":"barNum",  "y":svgHeight-5-offsetY,
               "x":function(d,i){return i * (barWidth + barMargin) + 10 + offsetX;}, })

    barElements.enter().append("text")              // 막대의 레이블을 표시
        .attr({"class":"barName",
               "x":function(d,i){return i*(barWidth + barMargin) + 10 + offsetX;}, })
        .transition().duration(500).delay(function(d,i){return i * 150;})                  // animation 설정
        .attr("y", svgHeight - offsetY + 15)
        .text(function(d,i){return ["A","B","C","D","E","F","G","H","I","J","K","L"][i];}) // 레이블 이름을 반환
    
    var yScale = d3.scale.linear().domain([0, dataMax]).range([dataMax, 0])  // 눈금표시 스케일
    d3.select("#myGraph_bar2").append("g")                                   // 세로 축 눈금(왼쪽)
        .call(d3.svg.axis().scale(yScale).orient("left"))
        .attr({"class":"axis", 
               "transform":"translate("+offsetX+", "+((svgHeight-300)-offsetY)+")",})
    d3.select("#myGraph_bar2").append("rect")                                // 가로 축 표시
        .attr({"class":"axis_x",  "width":svgWidth, "height":1,
               "transform":"translate("+offsetX+", "+(svgHeight-offsetY)+")"})
    </script>
</body>
```
